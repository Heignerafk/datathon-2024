```r
library(ggplot2)
library(ggplot2)
library(dplyr)

# Datos proporcionados
salario_minimo <- data.frame(
  ao = c(2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024),
  valor = c(737717, 781242, 828116, 877803, 908526, 1000000, 1160000, 1300000)
)

# Crear el grfico de lneas con degradado
ggplot(salario_minimo, aes(x = ao, y = valor)) +
  geom_line(aes(color = valor), size = 2) +
  scale_color_gradient(low = "darkgreen", high = "green") +
  geom_point(aes(color = valor), size = 5) +
  labs(
    title = "Incremento del salario mnimo en Colombia",
    x = "Ao",
    y = "Valor del salario mnimo (COP)",
    color = "Valor"
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 18, face = "bold"),
    axis.title = element_text(size = 14),
    axis.text = element_text(size = 12)
  )
```
